## Phần 1: Thống kê dữ liệu

### Sắp xếp điểm DH1 theo thứ tự tăng dần

```python
import pandas as pd

# Đọc file CSV
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Sắp xếp điểm DH1 theo thứ tự tăng dần
df_sorted_dh1 = df.sort_values(by='DH1')

# Cấu hình pandas để hiển thị toàn bộ dữ liệu
pd.set_option('display.max_rows', None)  # Hiển thị tất cả các hàng
pd.set_option('display.max_columns', None)  # Hiển thị tất cả các cột

# Hiển thị dữ liệu
print(df_sorted_dh1[['DH1']])

```

### Sắp xếp điểm DH2 theo nhóm giới tính tăng dần

```python
import pandas as pd

# Đọc file CSV
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Sắp xếp điểm DH2 theo nhóm GT
df_sorted_dh2 = df.sort_values(by=['GT', 'DH2'])

# Cấu hình pandas để hiển thị toàn bộ dữ liệu
pd.set_option('display.max_rows', None)  # Hiển thị tất cả các hàng
pd.set_option('display.max_columns', None)  # Hiển thị tất cả các cột

# Hiển thị dữ liệu đã sắp xếp
print(df_sorted_dh2[['GT', 'DH2']])

```

### Thống kê Q2 và Q3 của DH1 theo KT

```python
pivot_dh1_kt = df.pivot_table(values=['DH1'], index='KT', aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std'])
print(pivot_dh1_kt)
```

### Thống kê Q2 và Q3 của DH1 theo KT, KV và DT

```python
pivot_dh1_kt_kv_dt = df.pivot_table(values=['DH1'], index=['KT', 'KV', 'DT'], aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std'])
print(pivot_dh1_kt_kv_dt)
```

## Phần 2: Trình bày dữ liệu

### Lập bảng tần số và biểu đồ tần suất (cột và tròn)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Tính tần số
gt_freq = df['GT'].value_counts()

# Biểu đồ cột
sns.barplot(x=gt_freq.index, y=gt_freq.values)
plt.title('Tần số biến GT')
plt.xlabel('Giới tính')
plt.ylabel('Tần số')
plt.show()

# Biểu đồ tròn
gt_freq.plot(kind='pie', autopct='%1.1f%%')
plt.title('Tỷ lệ giới tính')
plt.ylabel('')
plt.show()
```

### Trình bày dữ liệu biến US_TBM1, US_TBM2, US_TBM3

```python
# Vẽ biểu đồ phân phối của từng biến
for var in ['US_TBM1', 'US_TBM2', 'US_TBM3']:
    sns.histplot(df[var], kde=True, bins=10)
    plt.title(f'Phân phối {var}')
    plt.xlabel(var)
    plt.ylabel('Tần số')
    plt.show()
```

### Lọc học sinh là nam

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc file CSV
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Lọc học sinh là nam
df_M = df[df['GT'] == 'Nam']

# Phân phối biến DT (Dân tộc) với học sinh là nam
dt_freq = df_M['GT'].value_counts()
print("Phân phối dân tộc của học sinh nam:")
print(dt_freq)
print(df.head())

```

```python
# Lọc học sinh nam thuộc dân tộc Kinh
df_filtered = df[
    (df['GT'] == 'Nam') &
    (df['DT'] == 'Kinh') &
    (df['DH1'] >= 5.0) &
    (df['DH2'] >= 4.0) &
    (df['DH3'] >= 4.0)
]

# Phân phối biến KV
kv_freq = df_filtered['KV'].value_counts()
print("Phân phối khu vực (KV) của học sinh nam thuộc dân tộc Kinh thỏa mãn điều kiện:")
print(kv_freq)
print(df.head())
```

### Hãy trình bày dữ liệu lần lượt các biến DH1, DH2, DH3 lớn hơn bằng 5.0 và thuộc khu
vực 2NT

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc file CSV
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Lọc dữ liệu với điều kiện DH1, DH2, DH3 >= 5.0 và thuộc khu vực 2NT
df_filtered = df[
    (df['DH1'] >= 5.0) &
    (df['DH2'] >= 5.0) &
    (df['DH3'] >= 5.0) &
    (df['KV'] == '2NT')
]

# Kiểm tra dữ liệu đã lọc
print("Dữ liệu đã lọc:")
print(df_filtered[['DH1', 'DH2', 'DH3', 'KV']])

# Trực quan hóa từng biến
plt.figure(figsize=(12, 8))

# Biến DH1
plt.subplot(3, 1, 1)
sns.histplot(df_filtered['DH1'], bins=10, kde=True, color='blue')
plt.title('Phân phối điểm DH1 (>= 5.0, KV = 2NT)')
plt.xlabel('Điểm DH1')
plt.ylabel('Tần số')

# Biến DH2
plt.subplot(3, 1, 2)
sns.histplot(df_filtered['DH2'], bins=10, kde=True, color='green')
plt.title('Phân phối điểm DH2 (>= 5.0, KV = 2NT)')
plt.xlabel('Điểm DH2')
plt.ylabel('Tần số')

# Biến DH3
plt.subplot(3, 1, 3)
sns.histplot(df_filtered['DH3'], bins=10, kde=True, color='red')
plt.title('Phân phối điểm DH3 (>= 5.0, KV = 2NT)')
plt.xlabel('Điểm DH3')
plt.ylabel('Tần số')

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
```
