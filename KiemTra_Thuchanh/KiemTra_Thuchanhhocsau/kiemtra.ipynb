```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Tạo một DataFrame từ dữ liệu 
data = {
    "Name": ["Mr.Harris", "Mrs.Bradley", "Miss.Laina", "Mrs.Heath", "Mr.William", "Mr.James", "Mr.Timothy", "Mrs.Oscar", "Mrs.Nicholas", "Miss.Marguerite", "Mr.William", "Miss.Hulda", "Mr.Lawrence", "Mr.William"],
    "Height": ["1.80", "1.61", "1.78", "1.74", "1.55", "1.9", "1.61", "1.40", "1.92", "1.77", "1.55", "1.85", "1.64", "1.77"],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "Age": [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
    
}
df = pd.DataFrame(data)
print("DataFrame ban đầu:")
print(df)
```

### Cau 14:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Tạo DataFrame với dữ liệu đã cho
data = {
    "Name": [
        "Mr. Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William",
        "Mr. James", "Mr. Timothy", "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite",
        "Mr. William", "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"
    ],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.90, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "Age": [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}

df = pd.DataFrame(data)
# Tính BMI
df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)

print(df)
```

### Cau 15:

```python
def classify_bmi(bmi):
    if bmi < 18.5:
        return 0  # Thiếu cân
    elif 18.5 <= bmi <= 22.9:
        return 1  # Bình thường
    elif 23 <= bmi <= 24.9:
        return 2  # Thừa cân
    else:
        return 3  # Béo phì

df["Phan_Loai"] = df["cs_BMI"].apply(classify_bmi)
print(df)
```

### Cau 16:

```python
# Chuẩn hóa cs_BMI về đoạn [0, 1]
df["cs_BMI"] = (df["cs_BMI"] - df["cs_BMI"].min()) / (df["cs_BMI"].max() - df["cs_BMI"].min())
print(df)
```

### Cau 17:

```python
df["GT"] = df["Name"].apply(lambda x: "male" if "Mr." in x else "female")

# Xóa tiền tố Mr, Mrs, Miss khỏi cột Name
df["Name"] = df["Name"].str.replace(r"Mr\.|Mrs\.|Miss\.", "", regex=True).str.strip()
print(df)
```

### Cau 18:

```python
# Đếm số lượng male và female
gender_counts = df["GT"].value_counts()

# Vẽ biểu đồ tròn
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%", colors=["blue", "pink"])
plt.title("Tỉ lệ giới tính")
plt.show()
```

### Cau 19:

```python
# Lọc dữ liệu nữ
df_female = df[df["GT"] == "female"]

# Vẽ biểu đồ cột số lượng nữ theo Phan_Loai
plt.figure(figsize=(6, 4))
df_female["Phan_Loai"].value_counts().sort_index().plot(kind="bar", color="purple")
plt.xticks(ticks=[0, 1, 2, 3], labels=["Thiếu cân", "Bình thường", "Thừa cân", "Béo phì"], rotation=0)
plt.ylabel("Số lượng")
plt.title("Phân loại BMI của nữ")
plt.show()

```
