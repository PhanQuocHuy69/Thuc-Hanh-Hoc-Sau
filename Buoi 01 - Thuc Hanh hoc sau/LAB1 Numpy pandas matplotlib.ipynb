## NUMPY

### 1.1 Tạo mảng với numpy

```python
import numpy as np
# Tạo mảng 1 chiều
mang_1_chieu = np.array([1,3,5,7,9])
print(mang_1_chieu)
# Tạo mảng 2 chiều
mang_2_chieu = np.array([[1,3,5,7,9],[2,4,6,8,10]])
print(mang_2_chieu)
# Tạo mảng ngẫu nhiên 
mang_ngau_nhien = np.random.rand(3,3)
print(mang_ngau_nhien)
```

### Các thao tác cơ bản của numpy

```python
mang_1_chieu = np.array([1,3,5,7,9])

# Cộng trừ nhân chia với một số
cong = mang_1_chieu + 2  # Cộng 2
tru = mang_1_chieu - 1   # Trừ 1
nhan = mang_1_chieu * 3  # Nhân 3
chia = mang_1_chieu / 2  # Chia 2

print("Cộng:", cong)
print("Trừ:", tru)
print("Nhân:", nhan)
print("Chia:", chia)

#Tính tổng, trung bình, lớn nhất, nhỏ nhất
tong = np.sum(mang_1_chieu)
trung_binh = np.mean(mang_1_chieu)
lon_nhat = np.max(mang_1_chieu)
nho_nhat = np.min(mang_1_chieu)

print("Tổng:", tong)
print("Trung bình:", trung_binh)
print("Lớn nhất:", lon_nhat)
print("Nhỏ nhất:", nho_nhat)
#Lấy phần tử bất kỳ trong mảng

phan_tu_dau = mang_1_chieu[0]   # Phần tử đầu tiên
phan_tu_cuoi = mang_1_chieu[-1] # Phần tử cuối cùng
phan_tu_thu_hai = mang_1_chieu[1]  # Phần tử thứ hai

print("Phần tử đầu:", phan_tu_dau)
print("Phần tử cuối:", phan_tu_cuoi)
print("Phần tử thứ hai:", phan_tu_thu_hai)
```

## PANDAS

```python
import pandas as pd

data = {
    "Ten": ["Khang", "Huy", "Kiet", "Bao", "Tin", "Khang Nguyen"],
    "Tuoi": [21, 21, 18, 18, 18, 18],
    "Diem": [10, 10, 8 , 9 ,7 ,8]
}

df = pd.DataFrame(data)
print(df)
```

```python
# In ra cột tuổi
print(df["Tuoi"])
# In ra hàng đầu tiên
print(df.iloc[0])
#In ra dữ liệu có điểm lớn hơn 8
print(df[df["Diem"] < 8])
#Tính trung bình điểm lớp học
print(df["Diem"].mean())
```

## Matplotlib

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ đường
x =[1,2,3,4,5]
y =[3,7,8,9,3]

plt.plot(x,y, label ="Bieu do duong", color ="red", marker ="o")
plt.title("Bieu do duong")
plt.xlabel("Truc x")
plt.ylabel("Truc y")
plt.legend()
plt.show
```

```python
# Bieu do cot

Ten =["Ly", "Mai", "Khang", "Huy", "An"]
Diem = [5,7,8,3,9]

plt.bar(Ten,Diem, color="yellow")
plt.title("Bieu do cot")
plt.xlabel("Ten")
plt.ylabel("Diem")
plt.show
```

```python
import numpy as np
# Bieu do phan tan

x = np.random.rand(100)
y = np.random.rand(100)

size = np.random.rand(100) * 100

plt.scatter(x,y, size, alpha = 0.5, color="green")
plt.title("Bieu do phan tan")
plt.xlabel("Truc x")
plt.ylabel("Truc y")
plt.show
```
